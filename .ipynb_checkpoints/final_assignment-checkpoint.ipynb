## FA.1 Convert all timeseries to Dfs0

```python
import os
import numpy as np
import pandas as pd
import mikeio

from mikeio import Dataset
from mikeio.eum import EUMType, ItemInfo, EUMUnit
```

```python
os.makedirs(r'.\results')
```

a) Convert all timeseries to dfs0 (remember that the notebook should be runnable for your peers so put the files somewhere reasonable). 

b) Read s15_east_novayork_river.dfs0, print the "header", plot, and show that the number of missing values is 0.

```python
# a)
path_in = r".\stream_data"
timeseries = [x for x in os.listdir(path_in) if x.endswith(".dat")]
```

```python
for timeserie in timeseries:
    df = pd.read_csv(os.path.join(path_in, timeserie), comment='%', index_col=0, sep='\t', parse_dates=True,na_values=-99)
    df = df.interpolate()
    ds = mikeio.from_pandas(df)
    timeserie = timeserie.split('.')[0]
    ds.to_dfs(rf'.\results\{timeserie}.dfs0')
```

## FA.2 Add region specific info to normalize timeseries with surface area

Each timeseries belongs to a region identified in the filename, e.g. `s15_east_novayork_river.dat` is located in the `novayork` region.

a) Convert all timeseries to dfs0 with specific discharge, by doing: 

For each timeseries in the dataset:
* Find out which region it belongs to (hint: the string method [split()](https://docs.python.org/3/library/stdtypes.html#str.split) will be useful)
* Divide the timeseries values with the surface area for the region (take into account units)
* Create a dfs0 file with specific discharge *(discharge / area)* (like the one with discharge from FA.1)

b) Determine which station has the largest max specific discharge (in the simulation period).  

```python
dfs0_files = os.listdir(r'./results_v2')

for dfs0_file in dfs0_files:
    region = dfs0_file.split('.')[0].split('_')[-2]
    print(region)
```

```python
region_info = pd.read_csv(r'.\stream_data\region_info.csv', index_col=0)
region_info = region_info.to_dict()
region_info
```

```python
specific_discharge_max = {}

for dfs0_file in dfs0_files:
    region = dfs0_file.split('.')[0].split('_')[-2]
    ds = mikeio.read(os.path.join(r'./results_v2', dfs0_file))
    
    ds.items[0].type = EUMType.Discharge
    ds.items[0].unit = EUMUnit.meter_pow_3_per_sec

    if region in region_info['surface_area_km2'].keys():
        ds_dis = (ds[0].values/region_info['surface_area_km2'][region])/1e6
        ds_new = mikeio.DataArray(data = ds_dis, time = ds.time, item=mikeio.ItemInfo(
                    "Specific Discharge",
                    mikeio.EUMType.Specific_Discharge,
                    mikeio.EUMUnit.meter_per_sec))
        
    ds_new.to_dfs(rf'.\results_v2\{dfs0_file.split('.')[0]}_specific_discharge.dfs0')

    if 'specific_discharge' in dfs0_file:
        station = dfs0_file.split('.')[0].split('_')[0]
        ds = mikeio.read(os.path.join(r'./results_v2', dfs0_file))
        specific_discharge_max[station] = ds[0].values.max()
        
```

```python
print(f'The maximum value of the specific discharge is: {max(specific_discharge_max.values())} m/s, and the region is {max(specific_discharge_max, key=specific_discharge_max.get)}')
```

ERA5 wave model for a part of the North Sea is available in the file `"../data/ERA5_DutchCoast.nc"`

```python
import xarray as xr
import mikeio
from mikeio import ItemInfo, EUMType, EUMUnit
import numpy as np

ds = xr.open_dataset("./ERA5_DutchCoast.nc")

ds
```

```python
lons = ds.longitude
lats_rev = ds.latitude[::-1] # reversed

# geometry = mikeio.Grid2D(x=...,y=..., projection="LONG/LAT")
#hm0 = mikeio.DataArray(np.flip(...,axis=1),geometry=geometry, time=ds.time, item=mikeio.ItemInfo(...)))
```

```python
# a)
variables = list(ds.keys())
variables
```

```python
# ds.mwd # Mean wave direction - degree
# ds.mwp # Mean Wave Period - s
# ds.mp2 # Mean zero-crossing wave period - s
# ds.pp1d # Peak wave period - s
# ds.swh # Significant height of combined wind waves and swell - m
# mikeio.ItemInfo(mikeio.EUMType.Mean_Wave_Direction)
EUMType.search('wave')
```

```python
# b) and c)
## SINCE THE LATITUDE VALUES WERE REVERSED, THE VALUES OF EACH ARRAY ALSO NEED TO BE REVERSED ALONG THE LATITUDE AXIS = 1! HENCE np.flip(..., axis=1)
g2D = mikeio.Grid2D(x=lons, y=lats_rev, projection="LONG/LAT")
xarrays = []
das = mikeio.Dataset([
    mikeio.DataArray(data=np.flip(ds.mwd.values, axis=1),
                     time=ds.time,
                     geometry=g2D,
                     item=mikeio.ItemInfo('mwd', mikeio.EUMType.Mean_Wave_Direction, mikeio.EUMUnit.degree)),
    mikeio.DataArray(data=np.flip(ds.mwp.values, axis=1),
                     time=ds.time,
                     geometry=g2D, 
                     item=mikeio.ItemInfo('mwp', mikeio.EUMType.Wave_period, mikeio.EUMUnit.second)),
    mikeio.DataArray(data=np.flip(ds.mp2.values, axis=1),
                     time=ds.time,
                     geometry=g2D, 
                     item=mikeio.ItemInfo('mp2', mikeio.EUMType.Undefined , mikeio.EUMUnit.second)),
    mikeio.DataArray(data=np.flip(ds.pp1d.values, axis=1),
                     time=ds.time,
                     geometry=g2D, 
                     item=mikeio.ItemInfo('pp1d', mikeio.EUMType.Undefined , mikeio.EUMUnit.second)),
    mikeio.DataArray(np.flip(ds.swh.values,axis=1),
                     time=ds.time,
                     geometry=g2D, 
                     item=mikeio.ItemInfo('swh', mikeio.EUMType.Significant_wave_height , mikeio.EUMUnit.meter))
])
das.to_dfs(r'.\results\dutch_coast.dfs2')
my_das = mikeio.read(r'.\results\dutch_coast.dfs2')
my_das['swh'].plot()
```

```python
# d)
# max_netcdf = (ds['swh'].sel(latitude=50.0, longitude=1.0)).max().values
max_netcdf = (ds.swh.sel(latitude=50.0, longitude=1.0).values).max()
max_dfs2 = (my_das['swh'].sel(x=1.0, y=50.0)).max().values

string ='The max swh from the NETCdf file is {:.4f}. \nThe max value from the dfs2 file is {:.4f}'.format(max_netcdf, max_dfs2)
print(string)
```
